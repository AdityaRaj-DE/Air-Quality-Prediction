```python
import pandas as pd

df = pd.read_csv("../data/city_day.csv")

print(df.shape)
print(df.columns)
df.head()
```

```python
df = df.drop(columns=["AQI_Bucket"])
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
df = df.dropna(subset=["AQI"])
```

```python
features = [
    "PM2.5", "PM10", "NO2", "SO2", "CO", "O3"
]

df = df[["City", "Date", "AQI"] + features]
```

```python
df[features] = df[features].fillna(df[features].mean())
```

```python
print(df.shape)
print(df.isnull().sum())
print(df.describe())
```

```python
df.to_csv("../data/clean_city_day_aqi.csv", index=False)
```
